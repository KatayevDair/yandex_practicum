### 1.2 проверка расчетов эффективности

Давайте проверим, правильно ли расчитана метрика recovery

```python
def recovery_func(row):
    c = row['rougher.output.concentrate_au']
    f = row['rougher.input.feed_au']
    t = row['rougher.output.tail_au']
    return ((c*(f-t))/(f*(c-t)))*100
```

```python
# data['recovery_re2'] = data.apply(recovery_func, axis = 1)
data_train['rougher.output.recovery.calculation'] = data_train.agg(recovery_func, axis = 1)
data_train['rougher.output.mae.calculation'] = mean_absolute_error(data_train['rougher.output.recovery'],
                                                                  data_train['rougher.output.recovery.calculation'],
                                                                  )
'MAE = {}'.format(data_train.loc[0,'rougher.output.mae.calculation'])
# data['recovery_re2'] = data.agg(recovery_func, axis = 0, args = ['rougher.output.concentrate_au', 
#                              'primary_cleaner.output.concentrate_au',
#                              'rougher.output.tail_au'])
```

-15 степень это очнь мало, можно сказать что recovery расчитано верно. а ошибка существует из-за окургления.

### 1.4 предобработка данных

```python
print('{:.2%} процентов пропсуков в обучающей выборке,\nпри размере равном {}'.format(data_train.isna().sum().sum()/data_train.shape[0],data_train.shape[0]))
```

как мы можем заметить, у нас слишком много пропусков в обучающей выборке. это может стать проблемой. и у нас есть три варианта развитий событий. либо удалить их, либо заполнять пропуски, либо выбрать модель, которая умеет работать с пропусками

в первом случае все ясно, просто удалить их. но чем он плох-тем, что у нас слишком много пропусков и не хочется терять такие ценные данные просто так.
но с другой стороны, у нас на руках останется 10 000 измерений, чего досаточно для корректной работы модели.

никогда не хочется заполнять пропуски потому-что невозможно соблюсти закон, на основе которого появляются данные в этом мире и заполнение, какое бы оно не было хорошее, неизбежно приведет к понижению точности

построить модель, которя будет учитывать пропуски тоже можно считать верным решением, но неизвестно, насколько лучше она себя покажет.

я предлагаю вот что, как мы условились до этого, что не будем использовать все признаки из тестовой выборки, чтобы не переобучить модель, так и поступим.сейчас мы проанализируем важность признаков, посмотрим, как они коррелируют между друг другом, выберем наиболее подходящие признаки. и в зависимости от распределения пропусков постараемся заполнить пропуски в наименее важных признаках и дропнем наблюдения, которые мы посчитали важными и в которых есть пропуски. я считаю, чо так будет поступить правильнее всего.
потом, на этапе построения модели, кроме всего прочего в конце мы еще проведем тест. возьмем лучшую модель и скормим ей данные, где мы дропнули пропуски, где мы их заполнили и сравним эти три модели + ту, что учитвает пропуски.

## Модель

для начала подготовим все. напишем функцию, которая будет оценивать работу модели машинного обучения и создадим множества правильных ответов target. есть три варианта, как мы можем сделать.  пытаться предсказать rougher.output.recovery и final.output.recovery по отдельности одной моделью и выситывать итоговый sMAPE, и третий,- сделать две модели для каждого признака. 
второй удобны для заказчика, потомо-что он просто будет скармливать алгоритму один датасет и не переживать. второй вариант не очень удобный, потому-что заказчику придется дополнительно разделять изначальный датасет на два и скармливать модели. но зато, возможно, второ1 вариант будет показывать разные результаты, потому-что, возможно, на предсказание этих двух величин по разному влияют признаки(их важность) и модель будет переобучена в обоих случаях, в если мы напишем два алгоритма, которые принимают на вход разные данные, возможно, такой проблемы не возникнет.

сначала мы проверим, какй подход из вышеуказынных даст лучший результат, и если между ними будет большая разница, то мы возьмем тот, у кого итоговая sMAPE выше. потом мы будем искать лучший алгоритм,подбирая гиперпараметры, сделаем сводную таблицу по всем моделям и выберем наиболее подходящую(ие) основываясь на показателях итоговой sMAPE(при  прочих равных выберем ту, у которой целевая метрика выше, а в случае равных или примерно равных целевых метрик будем смотреть на остальные.)

```python
def sMAPE(target, predict):
    smape = 0
    for i in range(len(target)):
        smape+=(abs(target[i]-predict[i])/((abs(target[i])+abs(predict[i]))/2))
    return smape*(100/(len(target)))
```

```python
def fin_sMAPE(rougher_smape, final_smape):
    return 0.25*rougher_smape+0.75*final_smape
```

я понял, что лучше будет написать функцию, которая считает все основные метрики

```python
def get_metrics(true_rougher, pred_rougher, true_final, pred_final):
    smape_rougher = sMAPE(pred_rougher, true_rougher)
    smape_final = sMAPE(pred_final, true_final)
    rmse_linear_rougher = math.sqrt(mean_squared_error(true_rougher,pred_rougher))
    rmse_linear_final = math.sqrt(mean_squared_error(true_final,pred_final))
    mae_linear_rougher = mean_absolute_error(true_rougher,pred_rougher)
    mae_linear_final = mean_absolute_error(true_final,pred_final)
    return rmse_linear_rougher, rmse_linear_final, mae_linear_rougher, mae_linear_final, fin_sMAPE(smape_rougher, smape_final)
```
